```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv(
    "https://gist.githubusercontent.com/saikrishnapotluri/33ace369025ec4de0dfb9f22a0c5b09f/raw/5d4d29e255ef0e7368e0808bd012fb3080ca22d7/salary_Data.csv"
)
df
```

```python
df.sort_values(by="YearsExperience", inplace=True)
```

```python
scaler = MinMaxScaler()
x, y = df["YearsExperience"].values.reshape(-1, 1), df["Salary"].values

x_scaled = scaler.fit_transform(x)

# train test split
x_train, x_test, y_train, y_test = train_test_split(
    x_scaled, y, test_size=0.2, random_state=42
)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
```

```python
plt.scatter(x_scaled, y, color="blue", label="Actual Data")
plt.plot(x_test, y_pred, color="red", label="Regression Line")
plt.xlabel("Years of Experience (Scaled)")
plt.ylabel("Salary")
plt.title("Linear Regression: Salary vs. Years of Experience")
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R2 score: {r2:.2f}")
```
